```python
import matplotlib.pyplot as plt
from matplotlib import font_manager

# Assuming we've already downloaded and added the XKCD font
font_path = "xkcd-script.ttf"
prop = font_manager.FontProperties(fname=font_path)

# Enable XKCD style
plt.xkcd()

# Create the figure
plt.figure(figsize=(8, 5))  # Increased figure size further

# Data
activities = ['Actually working', 'Pretending to work', 'Coffee breaks',
              'Staring blankly at screen', 'Fixing other people\'s code']
sizes = [20, 30, 15, 25, 10]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']
explode = (0, 0.1, 0, 0.1, 0)  # explode 2nd & 4th slices

# Create pie chart
wedges, texts, autotexts = plt.pie(sizes, explode=explode, labels=activities, colors=colors, autopct='%1.1f%%',
                                   startangle=90, pctdistance=0.75, labeldistance=1.1,
                                   wedgeprops=dict(width=0.5, edgecolor='white'))

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add title with larger font
plt.title('Where My Time Goes During a Workday', fontproperties=prop, y=1.08, fontsize=20)

# Customize font for labels and percentages
for text in texts + autotexts:
    text.set_fontproperties(prop)

# Set font size for labels
for text in texts:
    text.set_fontsize(18)

# Set font size for percentages
for autotext in autotexts:
    autotext.set_fontsize(16)
    autotext.set_weight('bold')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Create the figure
plt.figure(figsize=(8, 6))

# Data
hours = np.linspace(0, 24, 100)
productivity = 50 + 30 * np.sin(np.pi * hours / 12)
coffee = 20 + 50 * np.sin(np.pi * hours / 8 - 0.5)

# Create plots
plt.plot(hours, productivity, label='Productivity', linewidth=2)
plt.plot(hours, coffee, label='Coffee Level', linewidth=2)

# Customize the plot
plt.xlabel('Hour of the Day', fontproperties=prop, fontsize=16)
plt.ylabel('Level of Awesome', fontproperties=prop, fontsize=16)
plt.title('Productivity vs. Coffee Consumption', fontproperties=prop, fontsize=20)

# Set x-axis ticks
plt.xticks(np.arange(0, 25, 4), fontproperties=prop, fontsize=14)
plt.yticks(fontproperties=prop, fontsize=14)

# Add legend
plt.legend(prop=prop, fontsize=14)

# Add annotations
plt.annotate('Peak productivity!', xy=(6, 80), xytext=(10, 70),
             arrowprops=dict(arrowstyle='->'), fontproperties=prop, fontsize=18)
plt.annotate('Coffee crash :(', xy=(15, -25), xytext=(18, 15),
             arrowprops=dict(arrowstyle='->'), fontproperties=prop, fontsize=18)

plt.tight_layout()
plt.show()
```

```python
# Enable XKCD style
plt.xkcd()

# Create the figure
plt.figure(figsize=(14, 8))

# Data
categories = ['Actual coding', 'Googling errors', 'Staring at screen',
              'Explaining why it\'s "not a bug"', 'Pretending to understand the codebase']
values = [20, 35, 15, 20, 10]

# Create bar chart
bars = plt.bar(categories, values)

# Customize the plot
plt.ylabel('Time Spent (%)', fontproperties=prop, fontsize=18)
plt.title('Things I\'ve Learned in My Programming Career', fontproperties=prop, fontsize=20)

# Rotate and align the tick labels so they look better
plt.xticks(range(len(categories)), categories, fontproperties=prop, fontsize=18, rotation=15, ha='right')
plt.yticks(fontproperties=prop, fontsize=14)

# Add value labels on the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}%',
             ha='center', va='bottom', fontproperties=prop, fontsize=18)

# Add some space at the top of the plot for rotated labels
plt.subplots_adjust(top=0.9)

# Add a humorous annotation
plt.annotate('The truth hurts', xy=(1, 35), xytext=(2, 30),
             arrowprops=dict(arrowstyle='->'), fontproperties=prop, fontsize=18)

plt.tight_layout()
plt.show()
```

```python
# Enable XKCD style
plt.xkcd()

# Create the figure
plt.figure(figsize=(8, 6))

# Data
days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5']
x = np.arange(len(days))
y1 = np.array([100, 60, 30, 10, 0])  # Edible pizza
y2 = np.array([0, 30, 40, 30, 10])   # Questionable pizza
y3 = np.array([0, 10, 30, 60, 90])   # Pizza-shaped object

# Create stacked area chart
plt.fill_between(x, y1, label='Edible pizza', alpha=0.8)
plt.fill_between(x, y1, y1+y2, label='Questionable pizza', alpha=0.8)
plt.fill_between(x, y1+y2, y1+y2+y3, label='Pizza-shaped object', alpha=0.8)

# Customize the plot
plt.xlabel('Days in Fridge', fontproperties=prop, fontsize=16)
plt.ylabel('Pizza State (%)', fontproperties=prop, fontsize=16)
plt.title('The Life Cycle of a Pizza in My Fridge', fontproperties=prop, fontsize=20)

# Set x-axis ticks
plt.xticks(x, days, fontproperties=prop, fontsize=14)
plt.yticks(fontproperties=prop, fontsize=14)

# Add legend
plt.legend(prop=prop, fontsize=14, loc='center left')

# Add annotations
plt.annotate('Still delicious!', xy=(0, 90), xytext=(0.5, 95),
             arrowprops=dict(arrowstyle='->'), fontproperties=prop, fontsize=14)
plt.annotate('Might need a hazmat suit', xy=(4, 95), xytext=(3.5, 80),
             arrowprops=dict(arrowstyle='->'), fontproperties=prop, fontsize=14)

plt.tight_layout()
plt.show()
```

```python

# Enable XKCD style
plt.xkcd()

# Create the figure
plt.figure(figsize=(8, 6))

# Generate random data
np.random.seed(42)  # for reproducibility
time_choosing = np.random.randint(1, 121, 50)  # 1 to 120 minutes
enjoyment = 100 - (0.5 * time_choosing + np.random.randint(-20, 21, 50))
enjoyment = np.clip(enjoyment, 0, 100)  # Clip values to be between 0 and 100

# Create scatter plot
plt.scatter(time_choosing, enjoyment, s=100, alpha=0.7)

# Customize the plot
plt.xlabel('Time Spent Choosing a Movie (minutes)', fontproperties=prop, fontsize=16)
plt.ylabel('Movie Enjoyment (%)', fontproperties=prop, fontsize=16)
plt.title('The Netflix Paradox: Choice vs. Enjoyment', fontproperties=prop, fontsize=20)

# Set axis ticks
plt.xticks([0, 30, 60, 90, 120], fontproperties=prop, fontsize=14)
plt.yticks(range(0, 101, 20), fontproperties=prop, fontsize=14)

# Add a trend line
z = np.polyfit(time_choosing, enjoyment, 1)
p = np.poly1d(z)
plt.plot(time_choosing, p(time_choosing), "r--", alpha=0.8)


plt.tight_layout()
plt.show()
```
